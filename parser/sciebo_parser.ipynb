```python
import openpyxl

file_path = '231106_Möllmann_MedI_mRNA-Seq_5.xlsx'
wb = openpyxl.load_workbook(filename=file_path, data_only=True)

# Get All Sheets
a_sheet_names = wb.get_sheet_names()

# Get Sheet Object by names
o_sheet = wb.get_sheet_by_name('Overview')

# Sheet Maximum filled Rows and columns
print(o_sheet.max_row)
print(o_sheet.max_column)

for i in range(1,o_sheet.max_row):
    for j in range(1, o_sheet.max_column):
        o_cell = o_sheet.cell(row=i, column=j).value
        print(str(o_cell) + " - ", end ="")
    print("\n")


# o_cell = o_sheet.cell(row=2, column=1)
# print(o_cell.value)
# o_cell = o_sheet['H1']
# print(o_cell.value)
```

```python
file_path = '231106_Möllmann_MedI_mRNA-Seq_5.xlsx'

# Read the Excel file into a Pandas DataFrame
xl = pd.ExcelFile(file_path)


# Print sheet names
print("Sheet names:", xl.sheet_names)

# Iterate through each sheet and read data into a DataFrame
for sheet_name in xl.sheet_names:
    df = xl.parse(sheet_name)
    print(f"\nData from sheet: {sheet_name}")
    
    print(df)
```
